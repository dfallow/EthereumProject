```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## PLD
```

```python
pld = pd.read_csv('pld.csv')
pld
```

```python
maskPress = pld.loc[:,"# MaskPress.2s"]
press = pld.loc[:,"Press.2s"]
eprPress = pld.loc[:,"EprPress.2s"]
leak = pld.loc[:,"Leak.2s"]
respRate = pld.loc[:,"RespRate.2s"]
tidVol = pld.loc[:,"TidVol.2s"]
minVent = pld.loc[:,"MinVent.2s"]
snore = pld.loc[:,"Snore.2s"]
flowLim = pld.loc[:,"FlowLim.2s"]
crc = pld.loc[:,"Crc16"]
```

```python
maskPressPlot = plt.plot(maskPress.index, maskPress.values)
plt.ylabel("Mask pressure")
plt.xlabel("Measurements")
plt.savefig('../img/maskpress.png')
```

```python
pressPlot = plt.plot(press.index, press.values)
plt.ylabel("Machine pressure")
plt.xlabel("Measurements")
plt.savefig('../img/press.png')
```

```python
eprPressPlot = plt.plot(eprPress.index, eprPress.values)
plt.ylabel("Epr pressure")
plt.xlabel("Measurements")
plt.savefig('../img/eprpressure.png')
```

```python
leakPlot = plt.plot(leak.index, leak.values)
plt.ylabel("Leak rates")
plt.xlabel("Measurements")
plt.savefig('../img/leak.png')
```

```python
respRatePlot = plt.plot(respRate.index, respRate.values)
plt.ylabel("Respiration rate")
plt.xlabel("Measurements")
plt.savefig('../img/respiration.png')
```

```python
tidVolPlot = plt.plot(tidVol.index, tidVol.values)
plt.ylabel("Tidal volume")
plt.xlabel("Measurements")
plt.savefig('../img/tidvol.png')
```

```python
minVentPlot = plt.plot(minVent.index, minVent.values)
plt.ylabel("Min vent")
plt.xlabel("Measurements")
plt.savefig('../img/minvent.png')
```

```python
snorePlot = plt.plot(snore.index, snore.values)
plt.ylabel("Snore")
plt.xlabel("Measurements")
plt.savefig('../img/snore.png')
```

```python
flowLimPlot = plt.plot(flowLim.index, flowLim.values)
plt.ylabel("Flow lim")
plt.xlabel("Measurements")
plt.savefig('../img/flowlim.png')
```

```python
crcPlot = plt.plot(crc.index, crc.values)
plt.ylabel("Crc16")
plt.xlabel("Measurements")
plt.savefig('../img/crc.png')
```
