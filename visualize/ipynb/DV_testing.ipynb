## BRP

```python
brp = pd.read_csv('brp.csv')
```

```python
brp = brp.drop(['Crc16'], axis=1)
brp.rename(columns={'# Flow.40ms': 'Flow.40ms'}, inplace=True)
```

```python
brp.head()
```

```python
brp.shape
```

```python
# df = pd.DataFrame(custom_data)
```

```python
plt.title("Report from BRP")
plt.xlabel("Total process")
plt.ylabel(".40ms")
sns.lineplot(data=brp)
```

```python
sns.pairplot(brp, diag_kind='kde', markers="D")
```

```python
brp.plot(kind="scatter", x="Flow.40ms", y="Press.40ms")
```

```python
sns.jointplot(x="Flow.40ms", y="Press.40ms", data=brp, size=5)
```
